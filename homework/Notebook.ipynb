```python
train=pd.read_csv('../files/input/train_data.csv')
test=pd.read_csv('../files/input/test_data.csv')
```

```python
def clean_data(df):
    df=df.copy()
    df['Age']=2021-df['Year']
    if 'Year' and 'Car_Name' in df.columns:
        df=df.drop(columns=['Year','Car_Name'])
    return df

train=clean_data(train)
test=clean_data(test)
```
